```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_path='/Users/danielbabin/Documents/GitHub/U1474_XRF_Data_Report/Data/'
```

```python
MAD=pd.read_csv(data_path+'Carbonate_Allison.csv')
IODP=pd.read_csv(data_path+'Carbonates_20_11_2019.csv')
MAD.rename(columns={'Depth CCSF-361-U1474-ABDEF-20160220 (m)':'Depth CCSF (m)'},inplace=True)
IODP.rename(columns={'Top depth CCSF-361-U1474-ABDEF-20160220 (m)':'Depth CCSF (m)',
                     'Calcium carbonate (wt%)':'%CaCO3'},inplace=True)
```

```python
MAD['Depth CCSF (m)']=MAD['Depth CCSF (m)'].round(3)
IODP['Depth CCSF (m)']=IODP['Depth CCSF (m)'].round(3)
```

```python
IODP.columns
```

```python
MAD.columns
```

```python
MAD_col_list=['Exp', 'Site', 'Hole', 'Core', 'Type', 'Sect', 'A/W',
              'Depth CCSF (m)','%CaCO3','Code ']
```

```python
IODP_col_list=['Exp', 'Site', 'Hole', 'Core', 'Type', 'Sect', 'A/W',
              'Depth CCSF (m)','%CaCO3']
```

```python
both=pd.concat([MAD[MAD_col_list],IODP[IODP_col_list]],axis=0,sort=False)
both.set_index('Depth CCSF (m)',inplace=True)
both.sort_index(inplace=True)
```

```python
MAD.set_index('Depth CCSF (m)',inplace=True)
IODP.set_index('Depth CCSF (m)',inplace=True)
```

```python
both.loc[MAD.index,'Source']='This Study'
both.loc[IODP.index,'Source']='IODP'
```

```python
both['%CaCO3'].plot(marker='o',figsize=(16,8),ms=5)
```

# Drop Extremes
They were duplicated. This table shows

```python
dup_ids=both[both.index.duplicated()].index
both.loc[dup_ids]
```

```python
both=both[(both['%CaCO3']<60)&(both['%CaCO3']>15)]
```

# Duplicates
Take mean of duplicated

```python
dup_ids=both[both.index.duplicated()].index
```

```python
fig,ax=plt.subplots()
both.loc[dup_ids,'%CaCO3'].plot(marker='o',ls='',figsize=(16,8),ax=ax)
for i in dup_ids:
    ax.axvline(i)
```

## Mean

```python
mean_CaCO3=both.loc[dup_ids].groupby(by=both.loc[dup_ids].index).mean()
```

## Precision

```python
dups=both.loc[dup_ids]
```

```python
IODP_precision=(dups.loc[dups['Source']=='IODP','%CaCO3'].groupby(by=dups[dups['Source']=='IODP'].index).std()/
                dups.loc[dups['Source']=='IODP','%CaCO3'].groupby(by=dups[dups['Source']=='IODP'].index).mean())*100
```

```python
MAD_precision=(dups.loc[dups['Source']=='This Study','%CaCO3'].groupby(by=dups[dups['Source']=='This Study'].index).std()/
                dups.loc[dups['Source']=='This Study','%CaCO3'].groupby(by=dups[dups['Source']=='This Study'].index).mean())*100
```

```python
fig,ax=plt.subplots(figsize=(16,8))
both.loc[dup_ids,'%CaCO3'].plot(marker='o',ls='',ax=ax)
mean_CaCO3['%CaCO3'].plot(marker='o',ls='',ax=ax)
for i in dup_ids:
    ax.axvline(i)
```

## Drop Duplicates and Insert Average

```python
both.loc[dup_ids,'%CaCO3']=mean_CaCO3['%CaCO3']
```

```python
final=both[~both.index.duplicated(keep='first')].copy()
final.loc[final['Source']=='IODP','σ %CaCO3']=IODP_precision.mean()/100*final.loc[final['Source']=='IODP','%CaCO3']
final.loc[final['Source']=='This Study','σ %CaCO3']=MAD_precision.mean()/100*final.loc[final['Source']=='This Study','%CaCO3']
final=final[['Exp','Site','Hole','Core','Type','Sect','A/W','%CaCO3','σ %CaCO3','Source']]
```

```python
fig,ax=plt.subplots()
final['%CaCO3'].plot(yerr=final['σ %CaCO3'],capsize=1,ax=ax,marker='s',ms=3)
ax.set_ylabel('%CaCO$_3$')
```
